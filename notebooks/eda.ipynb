```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('../data/raw/data.csv')
```

```python
numeric_cols= ['Amount', 'Value']
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
df[['Amount', 'Value']].corr()
```

```python
df.isnull().sum()
```

```python
categorical_cols = ['ProductCategory', 'ProductId', 'ChannelId', 'PricingStrategy']

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

Insights
- Most transactions are small, with a few extreme outliers, the majority being 50K
- Aount and Value are strongly correlated (0.9)
- Money Spending is concentrated in a few categories: Financial Services, Airtime, Utility Bills, Data Bundles, and TV, while other categories are minor.
- Both numerical columns have extreme values
- No missing values, The dataset is clean
